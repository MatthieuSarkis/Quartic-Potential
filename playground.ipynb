```python
import numpy as np
print(list(np.linspace(0.5, 4.2, 100)))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('./plots.mplstyle')

with open('./results/Ar-Ar/r.npy', 'rb') as f:
    r = np.load(f)
with open('./results/Ar-Ar/epsilon.npy', 'rb') as f:
    epsilon = np.load(f)

r = r[epsilon > 0][r<=7.2]
epsilon = epsilon[epsilon > 0][r<=7.2]

fig, axes = plt.subplots(nrows=1, ncols=1)
axes.plot(r, epsilon)
axes.set_xlabel('Distance')
axes.set_ylabel('Ground state energy')
axes.grid(True)
plt.savefig('./assets/epsilon_vs_r_ArAr.png', dpi=300, transparent=False, bbox_inches='tight')
```
